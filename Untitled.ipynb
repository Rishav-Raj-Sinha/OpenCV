# Reading images using OpenCV
images are loaded using opencv and then stored as numpy arrays which alllows us to manipulate them 

```python
#read image in greayscale
img = cv2.imread("image.png",0) # here the second argument i.e 0 refers to the color code , 0 being the code for grayscale 
#print image as numpy 2D array 
print(img)
```

# Display the image attributes

```python
#print image size
print(img.shape)
#print data-type of image
print(img.dtype)
```

# Display image using matplotlib

```python
plt.imshow(img)
```

by default matplotlib uses a default colormap
if we want the iamge in greyscale then we need to specify the colormap

```python
plt.imshow(img,cmap='gray')
```

# Working with color images

```python
img = cv2.imread("image.png",1)
print(img.shape)
print(img.dtype)
```

here the 3 in (227, 242, 3) represents the 3 color channels thast the image ususes

```python
plt.imshow(img)
```

the colors are different because openCV uses BGR order of color therefore it passed BGR but matplotlib expected RGB

```python
#changing the order or colors
img_channel_reversed = img[:, :, ::-1]
plt.imshow(img_channel_reversed)
```

# splitting and merging color channels

- cv2.split() Divides a multi-channel array into several single-channel arrays.

- cv2.merge() Merges several arrays to make a single multi-channel array. All the input matrices must have the same size.
