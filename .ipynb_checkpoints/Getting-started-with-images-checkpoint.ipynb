```python
#read image in greayscale
img = cv2.imread("image.png",0) # here the second argument i.e 0 refers to the color code , 0 being the code for grayscale 
#print image as numpy 2D array 
print(img)
```

# Working with color images

```python
img = cv2.imread("image.png",1)
print(img.shape)
print(img.dtype)
```

here the 3 in (227, 242, 3) represents the 3 color channels thast the image ususes

```python
plt.imshow(img)
```

the colors are different because openCV uses BGR order of color therefore it passed BGR but matplotlib expected RGB

```python
#changing the order or colors
img_channel_reversed = img[:, :, ::-1]
plt.imshow(img_channel_reversed)
```

# splitting and merging color channels

- cv2.split() Divides a multi-channel array into several single-channel arrays.

- cv2.merge() Merges several arrays to make a single multi-channel array. All the input matrices must have the same size.


```python
img = cv2.imread("image.png",cv2.IMREAD_COLOR) #we can also use argument '1' instead of 'IMREAD_COLOR'
b, g, r = cv2.split(img)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r, cmap="gray");plt.title("Red Channel")
plt.subplot(142);plt.imshow(g, cmap="gray");plt.title("Green Channel")
plt.subplot(143);plt.imshow(b, cmap="gray");plt.title("Blue Channel")

# Merge the individual channels into a BGR image
imgMerged = cv2.merge((b, g, r))
# Show the merged output
plt.subplot(144)
plt.imshow(imgMerged[:, :, ::-1])
plt.title("Merged Output")
```

# changing colors

below we use the image imported in the previous shell (img), which was by default loaded in BGR color channels and we use builtin cv2 method to change the channels from BGR to RGB 

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

# Changing to HSV color space

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) #hue saturation vibrancy

# Split the image into the B,G,R components
h,s,v = cv2.split(img_hsv)

# Show the channels
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h, cmap="gray");plt.title("H Channel");
plt.subplot(142);plt.imshow(s, cmap="gray");plt.title("S Channel");
plt.subplot(143);plt.imshow(v, cmap="gray");plt.title("V Channel");
plt.subplot(144);plt.imshow(img);   plt.title("Original");
```

# Modifying individual values

```python
h_new = h+10

img_modified = cv2.merge((h_new, s, v )) # merges the old s and v values with new h value
img_modified_RGB = cv2.cvtColor(img_modified, cv2.COLOR_HSV2BGR)
plt.imshow(img_modified_RGB)

# The printed image will have a new hue and look visually different from the orginal image
```

# Writing/Saving images after manipulation

```python
img_BGR = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) # WE NEEDED TO DO THIS BECAUSE IF WE PASSED IMG(RGB) IT WILL BE SAVED AS BGR
cv2.imwrite("img_modified_RGB.png", img_BGR) # saves the image
Image(filename='img_modified_RGB.png')
```

the file-format of image will depend on the extension you give while passing the first argument to the 'imwrite' method
